```python
df = pd.read_csv('/Users/manu/Downloads/archive-2/mnist_train.csv')
```

```python
df.head(5)
```

```python
x = df.iloc[:, 1:]
y = df.iloc[:, 0]
```

```python
# Trimming down data as we need a small chunck of data

df1 = df.iloc[1: 100, :]
x1 = df1.iloc[:, 1:]
y1 = df1.iloc[:, 0]
```

```python
import seaborn as sns
```

```python
sns.heatmap(x1.iloc[5].values.reshape(28, 28))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators = 5)
```

```python
rfc.fit(x1, y1)
```

```python
rfc.feature_importances_
```

```python
sns.heatmap(rfc.feature_importances_.reshape(28, 28))
```

In the above graph we can witness that for MNIST data set the columns such as (0, 2), (2, 4), (22, 24), (24, 26) and rows such as (0, 2), (2, 4), (24, 26), (22, 24) are of less importance and hence data points around such can be removed and data similarly data points around the white block from heatmaps can be given prefrences/importance during interpretations etc.
